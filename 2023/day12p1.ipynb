```python
def parse_data(data):

    samples = [[int(x)-1 for x in list(line.split(' ')[0].replace('#', '2').replace('.', '1').replace('?', '0'))] for line in data]
    groups = [[int(x) for x in line.split(' ')[1].split(',')] for line in data]

    print(samples)
    print(groups)

    def check_sample_groups(sample, groups):
        string_sample = ''.join([str(x) for x in sample])
        sample_groups = string_sample.split('0')
        # remove empty strings
        sample_groups = [sum([int(y) for y in list(x)]) for x in sample_groups if x]

        if len(sample_groups) != len(groups):
            return False
        for i in range(len(sample_groups)):
            if sample_groups[i] != groups[i]:
                return False
        return True

    total = 0
    counter = 0
    for sample, group in zip(samples, groups):

        count_ones = sample.count(1)
        count_unknowns = sample.count(-1)
        
        group_total = sum(group)
        fill_ones = group_total - count_ones
        fill_zeros = count_unknowns - fill_ones

        where_unknowns = np.argwhere(np.array(sample) == -1).flatten()

        sample_total = 0
        for i in range(2**count_unknowns):
            choices = [int(x) for x in list(bin(i)[2:].zfill(count_unknowns))]

            if sum(choices) != fill_ones:
                continue
            if sum([1-x for x in choices]) != fill_zeros:
                continue
            
            new_sample = sample.copy()
            for j, choice in enumerate(choices):
                new_sample[where_unknowns[j]] = choice

            if check_sample_groups(new_sample, group):
                sample_total += 1

        # print(sample_total)
        total += sample_total

        counter += 1
        print(counter / len(samples) * 100, '%')

    total2 = 0

    print("part 1:", total)
    print("part 2:", total2)
```

```python
parse_data(input_data)
```
